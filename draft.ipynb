```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import folium
import plotly.express as px
from ortools.constraint_solver import pywrapcp, routing_ils_pb2
import osmnx as ox
import networkx as nx

import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('data_stores.csv')
```

```python
state = "DC"
city = "Washington"
country = "US"

df_regional = df[(df['State/Province']==state) & (df['City'] == city) & (df['Country']==country)][['Street Address',"Longitude","Latitude"]].sample(frac=1).reset_index(drop=True)
df_regional.head(3)
```

```python
df_regional = df_regional.drop_duplicates(keep='first')
```

```python
df_regional = df_regional.reset_index().rename(columns={'index':"id",'Longitude':'x','Latitude':'y'})
```

```python
df_regional.head(3)
```

```python
data = df_regional.copy()
```

```python
data['color'] = ''
data.loc[data["id"]==0,'color']='red'
data.loc[data["id"]!=0,'color'] = 'black'

start = data[data["id"]==0][['y','x']]
```

```python
data['color'].value_counts()
```

```python
#creating a map using folium

map = folium.Map(location=start ,titles = 'carthodbpositron', zoom_start=11)
data.apply(
    lambda row:
    folium.CircleMarker(
        location=[row['y'],row['x']],
        fill = True, color = row['color'], radius=4
    ).add_to(map),axis=1
)
map
```

```python
G = ox.graph_from_point(start, dist=100, network_type='drive')
G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)

fig,ax = ox.plot_graph(G,bgcolor="black", node_size = 30, node_color = 'green', figsize = (30,20))
```
